```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((500, 700, 3), dtype=np.uint8)


cv2.line(
    img,
    (50,50),       
    (300,50),       
    (0,255,0),      
    5               
)


cv2.rectangle(
    img,
    (50,100),       
    (250,250),      
    (255,0,0),     
    3
)


cv2.rectangle(
    img,
    (300,100),
    (500,250),
    (0,0,255),
    -1              
)


cv2.circle(
    img,
    (150,350),      
    70,             
    (0,255,255),
    4
)


cv2.circle(
    img,
    (400,350),
    10,
    (255,255,255),
    -1
)


cv2.putText(
    img,
    "OpenCV",
    (450,80),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255,255,255),
    2
)

points = np.array([
    [500,300],
    [600,300],
    [650,450],
    [550,450]
])

cv2.polylines(
    img,
    [points],
    True,
    (0,255,0),
    4
)


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(10,7))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
