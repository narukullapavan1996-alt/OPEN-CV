```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("puppy.jpg")

original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# Make all images same size
original = cv2.resize(original, (200, 200))
gray = cv2.resize(gray, (200, 200))
hsv = cv2.resize(hsv, (200, 200))
lab = cv2.resize(lab, (200, 200))

stacked = np.hstack((original, gray, hsv, lab))

plt.figure(figsize=(12,4))
plt.imshow(stacked)
plt.title("Original | Gray | HSV | LAB")
plt.axis("off")
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('puppy.jpg')

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
width, height = 300, 300
rgb = cv2.resize(rgb, (width, height))

hsv  = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
xyz  = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
yuv  = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)
lab  = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)

gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

result = np.hstack((rgb, hsv, xyz, gray_rgb, yuv, lab))
plt.figure(figsize=(24, 8))
plt.imshow(result)
plt.title("RGB | HSV | XYZ | GRAY | YUV | LAB", fontsize=16)
plt.axis('off')
plt.show()
```
