```python
import cv2
import pandas as pd

img = cv2.imread('puppy.jpg')

if img is None:
    print("Image not found")
else:
    

    pixels = img.reshape(-1, 3)

  
    df = pd.DataFrame(
        pixels,
        columns=['Blue', 'Green', 'Red']
    )

    
    df.to_excel(
        'pixel_values.xlsx',
        index=False
    )

    print("Pixel values saved to Excel")
```
